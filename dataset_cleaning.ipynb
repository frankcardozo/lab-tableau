```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#impoting dataset and analysing them
ds_products = pd.read_csv('./brazilian-ecommerce/olist_products_dataset.csv')
ds_items = pd.read_csv('./brazilian-ecommerce/olist_order_items_dataset.csv')
```

```python
# looking a shaape of datasets
ds_products.shape
```

```python
ds_items.shape
```

```python
# counting percentage of null values in each dataset
percent_missing = (ds_products.isnull().sum() / len(ds_products)) * 100
missing_value_ds_prod = pd.DataFrame({'percent_missing': percent_missing})
```

```python
missing_value_ds_prod.sort_values('percent_missing', inplace=True, ascending = False)
missing_value_ds_prod.head(10)
```

```python
percent_missing = (ds_items.isnull().sum() / len(ds_items)) * 100
missing_value_ds_items = pd.DataFrame({'percent_missing': percent_missing})
```

```python
missing_value_ds_items.sort_values('percent_missing', inplace=True, ascending = False)
missing_value_ds_items.head(10)
```

```python
# dropping columns that are not needed in each dataset
ds_prod = ds_products.drop(['product_name_lenght'],axis = 1)
ds_prod
```

```python
ds_items1 = ds_items.drop(['order_item_id', 'seller_id'],axis = 1)
ds_items1
```

```python
# Filtering items dataset
ds_items2 = ds_items1.pivot_table(index=["product_id"], values = ["price"])
ds_items2
```

```python
# Filtering items dataset
ds_items3 = ds_items1.pivot_table(index=["product_id"], values = ["freight_value"], aggfunc='mean') 
ds_items3
```

```python
# merging the two datasets on product id
merged = pd.merge(ds_items1, ds_prod, how="outer", on="product_id")
merged.head()
```

```python
# looking at null values after merging
percent_missing = (merged.isnull().sum() / len(merged)) * 100
missing_values = pd.DataFrame({'percent_missing': percent_missing})
missing_values.sort_values('percent_missing', inplace=True, ascending = False)
missing_values.head(5)
```

```python
# no problem with nulls
```

```python
#dropping order id column that is not going to be used
merged_clean = merged.drop(['order_id'], axis = 1)
```

```python
# calculating the volume of a product
merged_clean['product_volume'] = merged_clean['product_length_cm'] * merged_clean['product_height_cm'] * merged_clean['product_width_cm']
```

```python
# exporing merged and cleaned dataset to csv file
merged_clean.to_csv('merged_clean.csv', sep='\t', index=False)
```
